```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 1 
#Computing integral of e^(x^2) from 0 to 1 using trapezoidal rule 

#List containing the number of panels

N = [2,5,10,20,50,100,200,500,1000]
true_value = 1.4626517459071816
error_TR = []
error_TR_1 = [] #with first derivative endpoint corrections
error_TR_3 = [] #with third derivative endpoint corrections
h_vals = []

for n in N:
    panels = np.linspace(0,1,n+1)
    x_0 = panels[0]
    x_N = panels[-1]
    h = 1/(n)
    h_vals.append(h)
    TR = (h/2)*(np.e**(x_0**2)+np.e**(x_N**2))
    for i in range(1,len(panels)-1):
        TR += h*(np.e**(panels[i]**2))
        
    TR_1 = TR - ((h**2)/6)*np.e
    TR_3 = TR_1 +  ((h**4)/36)*np.e
    
    error_TR.append(np.abs(TR-true_value))
    error_TR_1.append(np.abs(TR_1-true_value))
    error_TR_3.append(np.abs(TR_3-true_value))
        
        
#plotting all the errors in a log log plot
plt.loglog(N,error_TR,'o-',label = 'error')
plt.loglog(N,np.asarray(h_vals)**2,label = "h^2")
plt.title("Loglog plot of the Absolute Error using Trapezoidal Rule",color = "white")
plt.grid(True)
plt.legend()
plt.ylabel(" |Error|",color = "white")
plt.xlabel("<-N->",color = "white")
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()

plt.loglog(N,error_TR_1,'o-',label = 'error')
plt.loglog(N,np.asarray(h_vals)**4,label = "h^4")
plt.title("Loglog plot of the Absolute Error using TR with first derivative endpoint correction",color = "white")
plt.grid(True)
plt.legend()
plt.ylabel(" |Error|",color = "white")
plt.xlabel("<-N->",color = "white")
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()

plt.loglog(N,error_TR_3,'o-',label = 'error')
plt.loglog(N,[h**6 for h in h_vals],label = "h^6")
plt.title("Loglog plot of the Absolute Error using TR with third derivative endpoint correction",color = "white")
plt.grid(True)
plt.legend()
plt.ylabel(" |Error|",color = "white")
plt.xlabel("<-N->",color = "white")
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()        
```

```python
# Question 3: Relative Error in n! using the Stirling formula

N = [20,50]

def factorial(n): #actual value of the n!
    if (n == 1):
        return 1
    else:
        return n*factorial(n-1)
    
def Stirling(n) :
    return ((n/np.e)**n)*np.sqrt(2*np.pi*n)

Rel_error = []

for n in N:
    Rel_error.append(np.abs(Stirling(n) - factorial(n))/factorial(n))
    
print(np.around(Rel_error,7))

    
```
